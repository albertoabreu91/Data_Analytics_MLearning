#### 1. Import pandas library.

```python
import pandas as pd
import numpy as np
```

#### 2. Import pymysql and sqlalchemy as you have learnt in the lesson of importing/exporting data. 


```python
from sqlalchemy import create_engine
import pymysql
```

#### 3. Create a mysql engine to set the connection to the server. Check the connection details in [this link](https://relational.fit.cvut.cz/dataset/Stats).

```python
driver = 'mysql+pymysql:'
user = 'guest'
password = 'relational'
ip = 'relational.fit.cvut.cz'
database= 'stats'

connection_string = f'{driver}//{user}:{password}@{ip}/{database}'
engine = create_engine(connection_string)
```

#### 4. Import the users table.

```python
query = """
SELECT *
FROM users
"""

users_df = pd.read_sql(query, engine)
```

```python
users_df.head()
```

#### 5. Rename Id column to userId.

```python
users_df = users_df.rename(columns = {'Id': 'userId'})
users_df.columns
```

#### 6. Import the posts table. 

```python
query = """
SELECT *
FROM posts
"""

posts_df = pd.read_sql(query, engine)
```

```python
posts_df.head()
```

#### 7. Rename Id column to postId and OwnerUserId to userId.

```python
posts_df = posts_df.rename(columns = {'Id': 'postId', 'OwnerUserId': 'userId'})
print(posts_df.columns)
print(users_df.columns)

users_df.head()
```

#### 8. Define new dataframes for users and posts with the following selected columns:
**users columns**: userId, Reputation, Views, UpVotes, DownVotes  
**posts columns**: postId, Score, userID, ViewCount, CommentCount

```python
users_sub_df = users_df[['userId', 'Reputation', 'Views', 'UpVotes', 'DownVotes']]

users_sub_df = users_sub_df.astype({'userId': float}) # Changing the type of the columns userId to float
```

```python
posts_sub_df = posts_df[['postId', 'Score', 'userId', 'ViewCount', 'CommentCount']]
```

#### 8. Merge both dataframes, users and posts. 
You will need to make a [merge](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.merge.html) of posts and users dataframes.

```python
# Merging user_sub_df with post_sub_df on column 'userId' as an Inner join.

user_posts_sub_merge = users_sub_df.merge(posts_sub_df, on = 'userId', how='inner')

user_posts_sub_merge.head()
```

#### 9. How many missing values do you have in your merged dataframe? On which columns?

```python
user_posts_sub_merge.info() # Cheking which columns have missing values
```

```python
'''
I have in the column ViewCount 48.396 NaN values
'''
```

#### 10. You will need to make something with missing values.  Will you clean or filling them? Explain. 
**Remember** to check the results of your code before going to the next step.

```python
user_posts_sub_merge['ViewCount'].value_counts(dropna=False).head() # Cheking how many NaN values I have.

#As there are 48.396 missing values, which counts for more than 50% of our data, it doesnt make sense to removed them
#And as there are no ViewCounts with 0 values, we can fill the NaN values with 0's.

user_posts_sub_merge = user_posts_sub_merge.fillna(value = 0)

user_posts_sub_merge.head() #Checking if the fillna method worked.

user_posts_sub_merge['ViewCount'].value_counts(dropna=False).head() #Double checking
```
